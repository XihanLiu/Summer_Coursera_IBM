```python
import pandas as pd
import numpy as np
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
concrete_data_columns = concrete_data.columns
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
n_cols = predictors.shape[1] # number of predictors
# build the model
model = regression_model()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
error_list_without_norm = []
for i in range(50):
  X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3)
  # fit the model
  model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, verbose=2)
  pred = model.predict(X_test)
  error_list_without_norm.append(mean_squared_error(pred, y_test))
```

```python
#Part A:
import statistics
print("Mean is {} \nStandard Deviation is {}".format(statistics.mean(error_list_without_norm), statistics.stdev(error_list_without_norm)))
'''
Mean is 54.90463343683415 
Standard Deviation is 26.91284281110331
'''
```

```python
concrete_data_columns = concrete_data.columns
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
predictors_norm = (predictors - predictors.mean()) / predictors.std()
n_cols = predictors_norm.shape[1] # number of predictors
# build the model
model = regression_model()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
error_list = []
for i in range(50):
  X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3)
  # fit the model
  model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, verbose=2)
  pred = model.predict(X_test)
  error_list.append(mean_squared_error(pred, y_test))
```

```python
#Part B:
import statistics
print("Mean is {} \nStandard Deviation is {}".format(statistics.mean(error_list), statistics.stdev(error_list)))
'''
Mean is 48.83850739285895 
Standard Deviation is 41.56315312215462
'''
```

```python
#Part C:
error_list_100_epochs = []
for i in range(50):
  X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3)
  # fit the model
  model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100, verbose=2)
  pred = model.predict(X_test)
  error_list_100_epochs.append(mean_squared_error(pred, y_test))

print("With 100 ephochs, Mean is {} \nStandard Deviation is {}".format(statistics.mean(error_list_100_epochs), statistics.stdev(error_list_100_epochs)))
'''
With 100 ephochs, Mean is 33.56115173911524 
Standard Deviation is 2.792850609553869
'''
```

```python
# define regression model
def regression_model_three_hidden():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```
